# Wandb

```python
import os

from pydub import AudioSegment
```

```python
base1 = '/kaggle/input/birdclef-2023-dataset/birdclef2021-background-noise/aicrowd2020_noise_30sec/noise_30sec'
base2 = '/kaggle/input/birdclef-2023-dataset/birdclef2021-background-noise/ff1010bird_nocall/nocall'
base3 = '/kaggle/input/birdclef-2023-dataset/birdclef2021-background-noise/train_soundscapes/nocall'
base_dirs = [base1, base2, base3]

out_base_dirs = ['/kaggle/input/birdclef-2023-dataset/birdclef2021-background-noise_wav/aicrowd2020_noise_30sec/noise_30sec',
                 '/kaggle/input/birdclef-2023-dataset/birdclef2021-background-noise_wav/ff1010bird_nocall/nocall',
                 '/kaggle/input/birdclef-2023-dataset/birdclef2021-background-noise_wav/train_soundscapes/nocall']
```

```python
# 各ディレクトリについて処理
for base_dir, out_base_dir in zip(base_dirs, out_base_dirs):
    for filename in os.listdir(base_dir):
        # .oggファイルのみを対象とする
        if filename.endswith(".ogg"):
            ogg_path = os.path.join(base_dir, filename)
            wav_path = os.path.join(out_base_dir, filename.replace(".ogg", ".wav"))
            
            # .oggをロードし.wavに変換
            ogg_audio = AudioSegment.from_ogg(ogg_path)
            ogg_audio.export(wav_path, format="wav")
            print(f"Converted {ogg_path} to {wav_path}")
```
